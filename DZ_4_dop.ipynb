```python
# Дополнительное задание 9
```

```python
x_corr.columns.values
```

```python
x_corr_abs = np.abs(x_corr)
x_corr_triu = x_corr_abs.where(~np.tril(np.ones(x_corr_abs.shape)).astype(np.bool))
x_corr_triu
```

```python
from collections import Counter
```

```python
high_corr = list(Counter(x_corr.columns[np.where(x_corr_triu >= 0.5)[0]]))
high_corr
```

```python
# Дополнительное задание 10
```

```python
x = x.drop('target', axis=1)
x.head()
```

```python
x_2 = x[high_corr]**2
x_2.head()
```

```python
x_2.rename(columns=lambda x: x+'_2', inplace=True)
x_2.head()
```

```python
x[x_2.columns.values] = x_2
x.head()
```
