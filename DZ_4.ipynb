```python
# Задание 1
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
data = boston.data
target = boston.target
feature_names = boston.feature_names
```

```python
data.shape
```

```python
feature_names
```

```python
x = pd.DataFrame(data, columns=feature_names)
x.head()
```

```python
y = pd.DataFrame(target, columns=['price'])
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
check_test = pd.DataFrame({'y_test': y_test['price'], 'y_pred': y_pred.flatten()}, columns=['y_test', 'y_pred'])
```

```python
check_test.head()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
# Задание 2
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(max_depth=12, n_estimators=1000, random_state=42)
```

```python
model.fit(x_train, y_train.values[:, 0])
```

```python
y_pred_forest = model.predict(x_test)
```

```python
check_test_forest = pd.DataFrame({'y_test': y_test['price'], 'y_pred': y_pred_forest.flatten()}, columns=['y_test', 'y_pred'])
```

```python
check_test_forest.head()
```

```python
r2_score(y_test, y_pred_forest)
```

```python
# 0.8749965273218174 > 0.7112260057484878, таким образом, алгоритм случайного леса работает в данном случае лучше
```
